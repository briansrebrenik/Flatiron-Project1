```python
import requests
from bs4 import BeautifulSoup 
```

```python
r = requests.get('https://www.ticketweb.com/event/action-bronson-white-warsaw-tickets/8883245?REFERRAL_ID=tmfeed')
c = r.content
soup = BeautifulSoup(c, 'html.parser')
soup
```

```python
price = soup.findAll('span', {'class': 'price theme-secondary-color small'})
output = float(price[0].text.replace('$', ""))
output
```
